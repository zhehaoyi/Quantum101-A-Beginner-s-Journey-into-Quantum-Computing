```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.patches import Rectangle, Circle

# ---- Style ----
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['font.size'] = 14
mpl.rcParams['axes.titleweight'] = 'bold'

# Colors (Nature-like palette)
colors = {
    "blue": "#377eb8",
    "red": "#e41a1c",
    "green": "#4daf4a",
    "orange": "#ff7f00",
    "gray": "#7f7f7f"
}

# 偏移距离 (arrow offset from object edge)
offset = 0.05

fig, axs = plt.subplots(1, 3, figsize=(16, 4.5))

# ===================== 1) First Law (Inertia) =====================
ax = axs[0]
ax.set_title("1. Inertia (First Law)")

# Ground lines
ax.plot([0.0, 1.2], [0, 0], lw=3, color=colors["gray"])
ax.plot([1.8, 3.0], [0, 0], lw=3, color=colors["gray"])

# Rest cart
cart1 = Rectangle((0.25, 0.0), 0.55, 0.28, color=colors["blue"], alpha=0.9)
ax.add_patch(cart1)
ax.text(0.525, 0.37, "At Rest", ha="center", va="center")

# Moving cart + velocity arrow (arrow outside cart, 5pt offset)
cart2_x, cart2_w = 1.90, 0.55
cart2 = Rectangle((cart2_x, 0.0), cart2_w, 0.28, color=colors["green"], alpha=0.9)
ax.add_patch(cart2)
start = (cart2_x + cart2_w + offset, 0.18)
end = (start[0] + 0.6, 0.18)
ax.annotate(
    "", xy=end, xytext=start,
    arrowprops=dict(arrowstyle="-|>", color=colors["red"], lw=3, mutation_scale=14)
)
ax.text(2.5, 0.37, "Constant velocity", ha="center", va="center")

ax.set_xlim(-0.1, 3.1)
ax.set_ylim(-0.15, 1.0)
ax.axis("off")

# ===================== 2) Second Law (F = ma) =====================
ax = axs[1]
ax.set_title("2. F = ma (Second Law)")

# Ground
ax.plot([0.0, 2.6], [0, 0], lw=3, color=colors["gray"])

# Mass block
block_x, block_w = 0.8, 1.0
block = Rectangle((block_x, 0.0), block_w, 0.5, color=colors["orange"], alpha=0.9)
ax.add_patch(block)
ax.text(block_x + block_w/2, 0.62, "Mass $m$", ha="center", va="center")

# External force arrow (offset left of block)
startF = (block_x - offset - 0.6, 0.25)
endF = (block_x - offset, 0.25)
ax.annotate("", xy=endF, xytext=startF,
            arrowprops=dict(arrowstyle="-|>", color=colors["red"], lw=3, mutation_scale=14))
ax.text(startF[0], 0.42, "Force $F$", color=colors["red"], ha="left")

# Acceleration arrow (offset right of block, same direction)
startA = (block_x + block_w + offset, 0.75)
endA = (startA[0] + 0.8, 0.75)
ax.annotate("", xy=endA, xytext=startA,
            arrowprops=dict(arrowstyle="-|>", color=colors["blue"], lw=3, mutation_scale=14))
ax.text((startA[0]+endA[0])/2, 0.92, "Acceleration $a$", color=colors["blue"], ha="center")

ax.set_xlim(-0.3, 2.8)
ax.set_ylim(-0.15, 1.2)
ax.axis("off")

# ===================== 3) Third Law (Action–Reaction) =====================
ax = axs[2]
ax.set_title("3. Action = −Reaction (Third Law)")

# Two persons aligned
A_x, B_x, y = 0.65, 2.25, 0.5
r = 0.23
A = Circle((A_x, y), r, color=colors["blue"], alpha=0.8)
B = Circle((B_x, y), r, color=colors["green"], alpha=0.8)
ax.add_patch(A); ax.add_patch(B)
ax.text(A_x, 0.17, "Person A", ha="center")
ax.text(B_x, 0.17, "Person B", ha="center")

# Action arrow (A → B), leaving gap from circles
start_act = (A_x + r + offset, y)
end_act = (B_x - r - offset, y)
ax.annotate("", xy=end_act, xytext=start_act,
            arrowprops=dict(arrowstyle="-|>", color=colors["red"], lw=3, mutation_scale=14))
ax.text((start_act[0]+end_act[0])/2, y+0.12, "Force", color=colors["red"], ha="center")

# Reaction arrow (B → A), same line, gap from circles
start_rea = (B_x - r - offset, y - 0.05)
end_rea = (A_x + r + offset, y - 0.05)
ax.annotate("", xy=end_rea, xytext=start_rea,
            arrowprops=dict(arrowstyle="-|>", color=colors["blue"], lw=3, mutation_scale=14))
ax.text((start_rea[0]+end_rea[0])/2, y-0.18, "Reaction", color=colors["blue"], ha="center")

ax.set_xlim(0.0, 2.9)
ax.set_ylim(0.0, 1.2)
ax.axis("off")

plt.tight_layout()

out_pdf = "newton_three_laws_nature_style_v3.pdf"
plt.savefig(out_pdf, bbox_inches="tight")
plt.show()

```
